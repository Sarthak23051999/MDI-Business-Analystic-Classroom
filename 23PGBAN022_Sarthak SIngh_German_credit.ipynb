```python
import matplotlib.pyplot  as plt
import seaborn as sns
%matplotlib inline
```

```python
def draw_cm(actual, predicted):
    cm = metrics.confusion_matrix(actual, predicted)
    sns.heatmap(cm, annot = True, fmt = '.2f', xticklabels = ["Good_credit", "Bad_Credit"], yticklabels = ["Good_credit", "Bad_credit"])
    plt.ylabel("True label")
    plt.xlabel("predicted label")
    plt.show()
```

```python
#ROC and AUC curve
#plot between Recall (Sensitivity) and 1-specificity. Different value of cutt off are used to get values for specificity and recall
def draw_roc(actual,probs):
    #obtain fpr, tpr, thresholds
    fpr,tpr,thresholds = metrics.roc_curve(actual,probs, drop_intermediate=False)
    auc_score = metrics.roc_auc_score(actual,probs)
    plt.figure(figsize=(8,6))
    plt.plot(fpr, tpr, label = "ROC Curve (area = %0.2f)" %auc_score)
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.05])
    plt.plot([0,1],[0,1], 'k--')
    plt.xlabel("False Positive rate or [1 - True Negative Rate]")
    plt.ylabel("True Positive rate")
    plt.legend(loc="lower right")
    plt.show()
    return fpr, tpr, thresholds
```

```python
fpr,tpr,thresholds = draw_roc(y_pred_df.actual, y_pred_df.predicted_prob)
```

```python
#Youden Index = specificity + sensitivity - 1 (sensitivity - 1 can be written as fpr)
#This is one of the methods through which we can get optimal threshold value 
tpr_fpr = pd.DataFrame({'tpr': tpr, 'fpr':fpr,'thresholds':thresholds})
tpr_fpr['diff'] = tpr_fpr['tpr']- tpr_fpr['fpr']
tpr_fpr.sort_values('diff',ascending = False)[0:5]
#therefore 0.221534 is chosen as best threshold
```
